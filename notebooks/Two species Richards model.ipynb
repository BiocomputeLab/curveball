## Schunte functions
Following [Schunte 1981](http://www.nrcresearchpress.com/doi/abs/10.1139/f81-153).

```python
def logistic(t, *params):
    A,mu,lam = params
    return A / ( 1 + exp(4*mu/A * (lam - t) + 2) ) 
def richards(t, *params):
    A,mu,lam,nu = params
    return A * (1 + nu * exp(1+nu+mu/A *(1+nu)**(1+1/nu) * (lam-t)))**(-1/nu)
def richards2(t, *params):
    a,b,y1,z1 = params
    t1 = min(t)
    return y1 * ( (a + b * z1 * (1 - exp(-a*(t-t1))))/a )**(1/b)
def logistic2(t, *params):
    a,y1,z1 = params
    b = -1
    return richards2(t, a,b,y1,z1)
def richards3(t, *params):
    K,alfa,nu,y0 = params
    t0 = min(t)
    Q = -1 + (K/y0)**nu
    return K / (1 + Q*exp(-alfa*nu*(t-t0)))**(1/nu)
def logistic3(t, *params):
    K,alfa,y0 = params
    nu = 1
    return richards3(t, K,alfa,nu,y0)
```

## Two species Richards model

$$
\frac{dy_i}{dt} = \alpha_i \cdot y_i \cdot \Bigg(1- \Big(\frac{\sum_j{y_j}}{K_i}\Big)^{\nu_i}\Bigg)
$$

```python
alfa1 = alfa
alfa2 = 0.9*alfa
K1,K2 = K,K
nu1,nu2 = nu,nu
def rde_2d(y,t):
    y1,y2 = y
    dy1dt = alfa1 * y1 * (1 - ((y1+y2)/K1)**nu1)
    dy2dt = alfa2 * y2 * (1 - ((y1+y2)/K2)**nu2)
    return dy1dt,dy2dt
```

```python
@interact
def plot_rde_2d(K=1000.,alfa=1.,nu=1.,y0=1.,T=30.):
    fig, (ax1,ax2) = subplots(2,1,figsize=(8,5))
    trange = arange(0, T)
    def _rde_2d(y,t):
        y1,y2 = y
        dy1dt = alfa * y1 * (1 - ((y1+y2)/K)**nu)
        dy2dt = y2 * (1 - ((y1+y2)/1000.)**1.)
        return dy1dt,dy2dt
    res = odeint(_rde_2d, (1.,1.), trange)
    yy1,yy2 = res.T
    ax1.plot(trange, yy1, 'k-')
    ax1.plot(trange, yy2, 'k--')
    ax2.plot(trange, gradient(yy1), 'k-')
    ax2.plot(trange, gradient(yy2), 'k--')
    ax1.set_xlabel('t')
    ax2.set_xlabel('t')
    ax1.set_ylabel('y')
    ax2.set_ylabel('dy/dt')
    ax1.set_ylim(0,1.1*K)
```

```python
trange = arange(min(t), max(t))
res = odeint(rde_2d, (y0,y0), trange)
```

```python
yy1,yy2 = res.T
plot(trange,yy1,'r--')
plot(trange,yy2,'b--')
#yscale('log')
axhline(y=y0, color='k', ls='--')
axhline(y=K, color='k', ls='--')
ylim(-0.1*K,1.1*K)
xlabel('t')
ylabel('OD');
```

```python
plot(trange[yy1+yy2<K/2],log(yy1/(yy2))[yy1+yy2<K/2], 'k')
xlabel('t')
ylabel('log (p/1-p)');
```

$$
s = \frac{d}{dt} \log{(\frac{y_1}{y2})} = \\
\frac{1}{\frac{y_1}{y2}} \cdot \frac{d(\frac{y_1}{y2})}{dt} = 
\frac{y_2}{y_1} \cdot \frac{\frac{dy_1}{dt} \cdot y_2 - y_1 \cdot \frac{dy_2}{dt} }{y_2^2} = \\
\frac{\frac{dy_1}{dt} \cdot y_2 - y_1 \cdot \frac{dy_2}{dt} }{y_1 y_2} = \\
\frac{\alpha_1 \cdot y_1 \cdot \Bigg(1- \Big(\frac{y_1 + y_2}{K_1}\Big)^{\nu_1}\Bigg) \cdot y_2 - y_1 \cdot \alpha_2 \cdot y_2 \cdot \Bigg(1- \Big(\frac{y_1 + y_2}{K_2}\Big)^{\nu_2}\Bigg) }{y_1 y_2} \Rightarrow \\
\alpha_1 \cdot \Bigg(1-\Big(\frac{y_1+y_2}{K_1}\Big)^{\nu_1}\Bigg) - \alpha_2 \cdot \Bigg(1-\Big(\frac{y_1+y_2}{K_2}\Big)^{\nu_2}\Bigg)
$$

```python
est_s = alfa1 * (1 - ((yy1+yy2)/K1)**nu1) - alfa2 * (1 - ((yy1+yy2)/K2)**nu2)
```

```python
plot(trange, gradient(log(yy1/yy2)), 'k')
plot(trange, est_s, '*')
xlabel('t')
ylabel('s')
axvline(x=max(trange[yy1+yy2<K/2]), color='k', ls='--');
s = gradient(log(yy1/(yy2))).max()
print s
```

## Stochastic Richards

The stochastic logistic function is ([Campillo et al. 2013](http://arxiv.org/abs/1307.2217)):
$$
dX_t = (\lambda - \mu - \alpha X_t)X_t dt + \rho \sqrt{(\lambda + \mu + \alpha X_t)X_t}dB_t
$$
