## Logistic fitting

```python
from scipy.optimize import curve_fit
```

$$
N(t) = \frac{K N_0 e^{rt}}{K + N_0 (e^{rt}-1)}
$$

```python
def logistic(t, *p):
    r, K, N0 = p
    return K * N0 * exp(r * t) / (K + N0 * (exp(r * t) - 1))
```

```python
(r_control,K_control,N0_control), pcov = curve_fit(logistic, repeat(t_control, y_control.shape[1]), y_control.flatten(), p0=(1,1,0))
perr = np.sqrt(np.diag(pcov))
print 'r',r_control,'K',K_control,'N0',N0_control
print 'errors',perr
```

```python
(r_assay,K_assay,N0_assay), pcov = curve_fit(logistic, repeat(t_assay, y_assay.shape[1]), y_assay.flatten(), p0=(1,1,0))
perr = np.sqrt(np.diag(pcov))
print 'r',r_assay,'K',K_assay,'N0',N0_assay
print 'errors',perr
```

```python
plot(t_control,  y_control, 'g', alpha=0.2)
plot(t_control, logistic(t_control, r_control, K_control, N0_control), 'k--', lw=2);
plot(t_assay,  y_assay, 'b', alpha=0.2)
plot(t_assay, logistic(t_assay, r_assay, K_assay, N0_assay), 'k--', lw=2);
```

## Logistic fit with diauxic phase

```python
start = 0
stop_assay = 100
stop_control = 100
t_assay = t[(start<t) & (t<stop_assay)]
y_assay = transpose(y[assay-1])[(start<t) & (t<stop_assay)]
plot(t_assay, y_assay, 'b', alpha=0.2);
y_control = transpose(y[control-1])[(start<t) & (t<stop_control)]
t_control = t[(start<t) & (t<stop_control)]
plot(t_control, y_control, 'g', alpha=0.2)
xlabel('time (hours)')
ylabel('OD 600');
```

$$
N(t) = 1_{t \le T} \cdot \frac{K N_0 e^{r_1 t}}{K + N_0 (e^{r_1 t}-1)} + 1_{t \ge T} \cdot N(T) + e^{r_2 (t-T)}
$$

```python
def logistic_diauxic(t, *p):
    r1, r2, K, N0, T, NT = p
    return (t<T) * K * N0 * exp(r1 * t) / (K + N0 * (exp(r1 * t) - 1)) + (t>=T) * K * NT * exp(r2 * t) / (K + NT * (exp(r2 * t) - 1))
```

```python
(r1_c, r2_c, K_c, N0_c, T_c, NT_c), pcov = curve_fit(logistic_diauxic, repeat(t_control, y_control.shape[1]), y_control.flatten(), 
                                                     p0=[(1,0.1,1,0,10,0.5)])
perr = np.sqrt(np.diag(pcov))
print 'r1',r1_c,'r2',r2_c,'K',K_c,'N0',N0_c,'T',T_c,'NT',NT_c
print 'errors',perr
```

```python
(r1_a, r2_a, K_a, N0_a, T_a, NT_a), pcov = curve_fit(logistic_diauxic, repeat(t_assay, y_assay.shape[1]), y_assay.flatten(), 
                                                     p0=[(1,0.1,K_c,N0_c,12.5,0.5)])
perr = np.sqrt(np.diag(pcov))
print 'r1',r1_a,'r2',r2_a,'K',K_a,'N0',N0_a,'T',T_a,'NT',NT_a
print 'errors',perr
```

```python
plot(t_control,  y_control, 'g', alpha=0.1)
plot(t_assay,  y_assay, 'b', alpha=0.1)
yscale('log')

t_ = linspace(t_control.min(),t_control.max())
plot(t_, [logistic_diauxic(t__, r1_c, r2_c, K_c, N0_c, T_c, NT_c) for t__ in t_], 'k--', lw=2);
t_ = linspace(t_assay.min(),t_assay.max())
plot(t_, [logistic_diauxic(t__, r1_a, r2_a, K_a, N0_a, T_a, NT_a) for t__ in t_], 'k--', lw=2);
```

```python
print "s1 =",log(2) * (r1_c-r1_a)/r1_a
print "s2 =",log(2) * (r2_c-r2_a)/r2_a
```

## Fitting modified Gompertz

$$
N(t) = A \cdot \exp{[-\exp{(1 + (\lambda-t) \frac{\mu \cdot e}{A})}]} 
+ A\cdot \exp{(\alpha(t-t_{shift}))}
$$

```python
def mod_gomp(t, *p):
    A,mu,lam,alpha,tshift = p
    return A * exp(-exp(1 + (lam - t) * mu*e/A)) + A * exp(alpha * (t-tshift))
```

```python
(A_c,mu_c,lam_c,alpha_c,tshift_c), pcov = curve_fit(mod_gomp, repeat(t_control, y_control.shape[1]), y_control.flatten(), 
                                                     p0=[(1,0.1,1,0.1,10)])
perr = np.sqrt(np.diag(pcov))
print 'A',A_c,'mu',mu_c,'lam',lam_c,'alpha',alpha_c,'shift',tshift_c
print 'errors',perr
```

```python
(A_a,mu_a,lam_a,alpha_a,tshift_a), pcov = curve_fit(mod_gomp, repeat(t_assay, y_assay.shape[1]), y_assay.flatten(), 
                                                     p0=[(1,0.1,1,0.1,10)])
perr = np.sqrt(np.diag(pcov))
print 'A',A_a,'mu',mu_a,'lam',lam_a,'alpha',alpha_a,'shift',tshift_a
print 'errors',perr
```

```python
plot(t_control,  y_control, 'g', alpha=0.1)
t_ = linspace(t_control.min(),t_control.max())
plot(t_, [mod_gomp(t__, A_c,mu_c,lam_c,alpha_c,tshift_c) for t__ in t_], 'k--', lw=2);

plot(t_assay,  y_assay, 'b', alpha=0.1)
t_ = linspace(t_assay.min(),t_assay.max())
plot(t_, [mod_gomp(t__, A_a,mu_a,lam_a,alpha_a,tshift_a) for t__ in t_], 'k--', lw=2);

yscale('log');
```

## Fitting logistic with lag phase

$$
N(t) = \frac{A}{1 + \exp(2 + (\lambda - t)\frac{4\mu}{A})}
$$

```python
def logistic2(t, *p):
    A, mu, lam = p
    return A / (1 + exp(2 + (lam - t) * 4 * mu / A))
```

```python
start = 0
stop_assay = 13
stop_control = 11
t_assay = t[(start<t) & (t<stop_assay)]
y_assay = transpose(y[assay-1])[(start<t) & (t<stop_assay)]
plot(t_assay, y_assay, 'b', alpha=0.2);
y_control = transpose(y[control-1])[(start<t) & (t<stop_control)]
t_control = t[(start<t) & (t<stop_control)]
plot(t_control, y_control, 'g', alpha=0.2)
xlabel('time (hours)')
ylabel('OD 600');
```

```python
(A_c,mu_c,lam_c), pcov = curve_fit(logistic2, repeat(t_control, y_control.shape[1]), y_control.flatten(), 
                                                     p0=[(1,0,0)])
perr = np.sqrt(np.diag(pcov))
print 'A',A_c,'mu',mu_c,'lam',lam_c
print 'errors',perr
```

```python
(A_a,mu_a,lam_a), pcov = curve_fit(logistic2, repeat(t_assay, y_assay.shape[1]), y_assay.flatten(), 
                                                     p0=[(1,0,0)])
perr = np.sqrt(np.diag(pcov))
print 'A',A_a,'mu',mu_a,'lam',lam_a
print 'errors',perr
```

```python
plot(t_control,  y_control, 'g', alpha=0.1)
t_ = linspace(t_control.min(),t_control.max())
plot(t_, [logistic2(t__, A_c,mu_c,lam_c) for t__ in t_], 'k--', lw=2);

plot(t_assay,  y_assay, 'b', alpha=0.1)
t_ = linspace(t_assay.min(),t_assay.max())
plot(t_, [logistic2(t__, A_a,mu_a,lam_a) for t__ in t_], 'k--', lw=2);

yscale('log');
```

## Richards

$$
N(t) = A \cdot [ 1 + \nu \cdot exp(1 + \nu + \frac{\mu}{A} \cdot (1+\nu)^{1+1/\nu} \cdot (\lambda-t))]^{-1/\nu}
$$


```python
def richards(t, *p):
    A,mu,lam,nu = p
    return A * (1. + nu * exp(1. + nu + (mu/A) * ((1. + nu)**(1. + 1./nu)) * (lam - t)))**(-1./nu)
```

```python
(A_c,mu_c,lam_c,nu_c), pcov = curve_fit(richards, repeat(t_control, y_control.shape[1]), y_control.flatten(), 
                                                     p0=[(1,0,0,1)])
perr = np.sqrt(np.diag(pcov))
print 'A',A_c,'mu',mu_c,'lam',lam_c,'nu',nu_c
print 'errors',perr
```

```python
(A_a,mu_a,lam_a,nu_a), pcov = curve_fit(richards, repeat(t_assay, y_assay.shape[1]), y_assay.flatten(), 
                                                     p0=[(1,0,0,1)])
perr = np.sqrt(np.diag(pcov))
print 'A',A_a,'mu',mu_a,'lam',lam_a,'nu',nu_a
print 'errors',perr
```

```python
plot(t_control,  y_control, 'g', alpha=0.1)
t_ = linspace(t_control.min(),t_control.max())
plot(t_, [richards(t__, A_c,mu_c,lam_c,nu_c) for t__ in t_], 'k--', lw=2);

plot(t_assay,  y_assay, 'b', alpha=0.1)
t_ = linspace(t_assay.min(),t_assay.max())
plot(t_, [richards(t__, A_a,mu_a,lam_a,nu_a) for t__ in t_], 'k--', lw=2);

yscale('log');
```
