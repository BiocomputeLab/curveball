```python
%pylab inline
import pandas as pd
import seaborn as sns
sns.set_style('ticks')
figsize(8,6)
```

```python
green,mixed,red,blank = 'green','mixed','red','blank'

def well2strain(well):    
    num = int(well[1:])
    if well == 'H12' or well == 'G12':
        return blank
    if 1<= num <= 4:
        return green
    if 5<= num <= 8:
        return mixed
    if 9<= num <= 12:
        return red
    
def OVERtoNaN(x):
    if x == 'OVER':
        return np.nan
    return x
```

```python
def read_tecan_tsv(filename, label):
    with open(filename) as f:
        for line in f:
            if line.startswith(label):
                break
        data = {}
        for line in f:
            # key
            if len(line.strip()) == 0:
                break
            line = line.strip().split('\t')
            if line[0].startswith('Cycle'):
                continue# key = 'Cycle'
            elif line[0].startswith('Temp'):
                continue# key = 'Temp'
            elif line[0].startswith('Time'):
                key = 'Time'
            else:
                key = line[0]
            # values
            values = map(OVERtoNaN, line[1:])
            
            values = map(float64, values)
            if key == 'Time':
                values = [v/60 for v in values]
            
            data[key] = values
            
    l = min([len(v) for v in data.values()])
    for k,v in data.items():
        data[k] = v[:l]
    df = pd.DataFrame(data)
    df = pd.melt(df, id_vars='Time', var_name='Well', value_name=label)
    df['Letter'] = map(lambda x: x[0], df.Well)
    df['Number'] = map(lambda x: int(x[1:]), df.Well)
    df['Strain'] = map(well2strain, df.Well)
    return df
df = read_tecan_tsv(filename, 'OD')
```

```python
def plot_96wells(df, label):
    g = sns.FacetGrid(df, hue='Strain', row='Number', col='Letter', size=3, 
                      hue_order=[red,mixed,green,blank], palette=sns.color_palette("Set1", 4))
    g.map(plot, "Time", label)
    return g
g = plot_96wells(df, 'OD')
```

## Logistic and Richards models

```python
def richards(t, *params):
    K,alfa,nu,y0 = params
    t0 = np.min(t)
    Q = -1 + (K/y0)**nu
    return K / (1 + Q * np.exp(-alfa*nu*(t-t0)))**(1/nu)
def logistic(t, *params):
    K,alfa,y0 = params
    nu = 1
    return richards(t, K,alfa,nu,y0)
```

```python
from scipy.optimize import curve_fit
def fit(df,t,obs,func,p0,PLOT=False):
    t_ = df[t]
    obs_ = df[obs]
    params, covars = curve_fit(func, t_, obs_, p0, maxfev=100000)    
    if PLOT:
        est = func(t_, *params)
        #residuals = obs - est
        #ssres = np.sum(residuals**2)
        plot(t_,est,label=func.__name__)
        scatter(t_,obs_, marker='.')
        xlabel(t)
        ylabel(obs)
    return params
```

```python
K,alfa,y0 = fit(df[(df['Well'] == 'A1') & (df['Time'] < 500)], 'Time', 'OD', logistic, (1,0.1,0), False)
print "K=%.3f alfa=%.3f OD(0)=%.3f" % (K,alfa,y0)
K,alfa,nu,y0 = fit(df[(df['Well'] == 'A1') & (df['Time'] < 500)], 'Time', 'OD', richards, (K,alfa,1,y0), False)
print "K=%.3f alfa=%.3f nu=%.3f OD(0)=%.3f" % (K,alfa,nu,y0)
#legend(loc='upper left', fontsize=16)
#title('A1');
```

```python
condition = (df['Strain'] == 'green')  & (df['Time'] < 500)
condition = condition & ~((df['Letter']=='D') | (df['Letter']=='E') | (df['Letter']=='F') | (df['Letter']=='G'))
K,alfa,y0 = fit(df[condition].sort('Time'), 'Time', 'OD', logistic, (1,0.1,0), True)
print "K=%.3f alfa=%.3f OD(0)=%.3f" % (K,alfa,y0)
K,alfa,nu,y0 = fit(df[condition].sort('Time'), 'Time', 'OD', richards, (K,alfa,1,y0), True)
print "K=%.3f alfa=%.3f nu=%.3f OD(0)=%.3f" % (K,alfa,nu,y0)
legend(loc='upper left', fontsize=16)
title('GFP')
axhline(y=K, color='k',ls='--',alpha=0.5)
axhline(y=y0, color='k',ls='--',alpha=0.5);
```
